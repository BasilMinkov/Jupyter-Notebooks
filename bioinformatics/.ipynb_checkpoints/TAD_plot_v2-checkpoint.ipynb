```python
from os.path import exists
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('ticks')
```

## Matrix inisialization

```python
matrix_path = './Simulations/simHiC_countMatrix_1227180_10-26_18h34m41s.txt'

matrix = np.matrix(pd.read_csv(matrix_path,
                               sep='\s',
                               index_col=None,
                               header=None))
```

### Searching for order of importance

```python
tads_path = '../From PC/TADtree_out/s50m10p7q7g500n150/simulated12_1' #/N49.txt
TADs = pd.DataFrame(columns=['chr', 'start', 'end'])
ok = True
c = 0
while ok:
    c += 1
    name = tads_path + '/N' + str(c) + '.txt'
    if exists(name):
        TADsN = pd.read_csv(name, sep='\t')
        TAD = pd.concat([TADsN, TADs, TADs]).drop_duplicates(keep=False)
        TADs = TADs.append(TAD, ignore_index = True)
    else:
        ok = False
```

## Visualization

```python
matrix.shape
```

```python
startmtx = 2400
endmtx = 3000
    

def mtxplot(TAD, figname=None, startmtx=0, endmtx=-1):
    nTADs = TAD.shape[0]
    for i, j in enumerate(TAD.index):
        bgn = TAD.loc[j].start - startmtx
        end = TAD.loc[j].end - startmtx
        if startmtx < bgn or end < endmtx: 
            plt.plot([bgn, end], [bgn, bgn], color=(i/nTADs, 0, 1-i/nTADs), linewidth=3.0)
            plt.plot([end, end], [bgn, end], color=(i/nTADs, 0, 1-i/nTADs), linewidth=3.0)
    plt.savefig(figname, dpi=200)
        
plt.figure(figsize=(20,20))
sns.heatmap(matrix[startmtx:endmtx, startmtx:endmtx], cmap='Greys')
mtxplot(TADs, figname='New_line3_S50N150_'+str(startmtx)+'_'+str(endmtx)+'.jpg', startmtx=startmtx, endmtx=endmtx)
```
