```python
import os
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)


from Bio import SeqIO


def get_protein_lengths(filename):
    
    protein_length = []
    
    with open(filename, "rU") as handle:
        for record in SeqIO.parse(handle, "fasta"):
            print(record.seq)
            print("\n")
            protein_length.append(len(record.seq))
    
    return protein_length

fastas = list(filter(lambda x: x.split(".")[-1] == 'fasta', os.listdir()))

protein_lengths = [get_protein_lengths(fasta) for fasta in fastas]

# plt.hist(protein_lengths)
```
