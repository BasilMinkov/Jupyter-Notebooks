```python
data = pd.DataFrame({"Time (ms)":[], "Trace #1":[], "State":[], "Cell":[]})

for i, file in enumerate(files, 1):
    for j in np.arange(1, 5):
        temp = pd.read_excel(io=path+file, sheet_name="Лист{}".format(j), parse_cols=[0,1])
        temp["Cell"] = i
        temp["State"] = j
        data = pd.concat([data, temp])
```

```python
x = data[(data["Cell"] == 1) & (data["State"] == 1)]
peaks, _ = find_peaks(x["Trace #1"], height=0)
plt.plot(x["Trace #1"])
plt.plot(peaks, x["Trace #1"][peaks], "x")
plt.plot(np.zeros_like(x["Trace #1"]), "--", color="gray")
plt.show()
```

```python
def highlight_local_max(x):  
    color = np.zeros(x.shape[0])

    for i, logical in enumerate(thr):
            if logical:
                color[i] = i 
            else:
                color[i] = NAN    
    plt.fill_between(color[o:n], np.ones(color.shape[0])[o:n]*40, np.ones(color.shape[0])[o:n]*-40, facecolor="red")


def find_spikes(x, peaks):

    thr = x["Trace #1"].diff().abs() > 1
    enter = []
    exit = []

    en_b = 0
    ex_b = 0
    peak_id = 0
    
    for i in np.arange(1, x.shape[0]-1):
        if not thr[i-1] and thr[i] and thr[i+1] and x["Trace #1"][i] < 0:
            en_b = i
        if thr[i-1] and thr[i] and not thr[i+1] and x["Trace #1"][i] < 0:
            ex_b = i
            if en_b < peaks[peak_id] and peaks[peak_id] < ex_b:
                enter.append(en_b)
                exit.append(ex_b)
                peak_id += 1
                if peak_id >= peaks.shape[0]:
                    return np.array(enter), np.array(exit)
                
    return np.array(enter), np.array(exit)

enter, exit = find_spikes(x, peaks)

def print_stat(x, enter, peaks, exit)
    print("Average spike frequency:", end=" ")
    print("{} s".format(round(x["Time (ms)"][peaks].diff().mean()*1e-3, 3)))
    print("Average spike length:", end=" ")
    print("{} ms".format(round((x["Time (ms)"][exit].values-x["Time (ms)"][enter].values).mean(), 3)))
    print("Average depolarisation length:", end=" ")
    print("{} ms".format(round((x["Time (ms)"][peaks].values-x["Time (ms)"][enter].values).mean(), 3)))
    print("Average repolarization length:", end=" ")
    print("{} ms".format(round((x["Time (ms)"][exit].values-x["Time (ms)"][peaks].values).mean(), 3)))
    print("Average amplitude change during depolarisation:", end=" ")
    print("{} mV".format(round((x["Trace #1"][peaks].values-x["Trace #1"][enter].values).mean(), 3)))
    print("Average spike power:", end=" ")
    print("{} W".format(np.mean([((x["Trace #1"][enter[i]:exit[i]]*1e-3)**2).sum() for i in range(len(enter))])))
```
