```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt


class QuantileSmoother:
    def __init__(self, n_samples, q):
        self.n_samples = n_samples
        self.buffer = np.zeros((n_samples,))
        self.q = q
        
    def apply(self, chunk: np.ndarray):
        chunk_size = chunk.shape[0]
        self.chunk_size = chunk_size
        if chunk_size <= self.n_samples:
            self.buffer[:-chunk_size] = self.buffer[chunk_size:]
            self.buffer[-chunk_size:] = chunk
        else:
            self.buffer = chunk[-self.n_samples:]
        return np.ones(chunk_size)*np.percentile(self.buffer, self.q*100)
```

```python
# generate data
fs = 500
x = np.random.normal(size=fs*60)
x = filtfilt(*butter(4, 2/fs*2), x)
```

```python
# init smooter
sm = QuantileSmoother(fs*10, 0.92)
```

```python
# simulate exp
y = []
chank_size = 8
t0 = 0
n_spindles = 0
counter = 0
for k in range(0, len(x), chank_size):
    counter += chank_size
    threshold = sm.apply(x[k:k+chank_size])
    y.append(threshold)
    if counter-t0 > 3*fs and x[k:k+chank_size].max() > threshold[-1]:
        n_spindles += 1
        t0 = counter
    
y = np.concatenate(y)
```

```python
# viz
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
print('#spindles =', n_spindles)
```
