# Интерполяция кубическими сплайнами и метод прогонки

## Интерполяция кубическими сплайнами

Интерполяция кубическим сплайном - это форма интерполяции, где интерполянт - это особый тип кусочно-заданных функций, называемых сплайнами. Интерполяция сплайнами часто лучше приближает функцию, чем интерполяция полиномами высоких степеней, устраняет проблему феномен Рунге – эффект нежелательных осцилляций, при котором колебания могут возникать между дискретными значениями функции.

Кубический сплайн дефекта $1$ – функция $S(x)$, которая:
1. На каждом отрезке $[x_{i-1},x_{i}]$ является многочленом степени не выше третьей.
2. Имеет непрерывные первую и вторую производные на всём отрезке $[a,b]$.
3. В точках $x_{i}$ выполняется равенство $S(x_{i})=f(x_{i})$.

В таком случае для $n+1$ узлов ${\left\{(x_{i},y_{i}):i=0,1,\cdots ,n\right\}}$ необходимо инторполировать каждые два узла $(x_{i},y_{i})$ и $(x_{i+1},y_{i+1})$ полиномами $y=q_{i}(x),i=1,2,\cdots ,n$.

Из-за необходимости минимизировать изгиб функции, и $y',, и $y ''$ должны быть неприрывными. Отсюда следует условие: 

${\begin{cases}q'_{i}(x_{i})=q'_{{i+1}}(x_{i})\\q''_{i}(x_{i})=q''_{{i+1}}(x_{i})\end{cases}}\qquad 1\leq i\leq n-1$

Условие удовлетворяется для полиномов третьей степени. 

Если представить полином третьей степени в симметрическом виде 

${\displaystyle q(x)=(1-t(x))y_{1}+t(x)y_{2}+t(x)(1-t(x))(a(1-t(x))+bt(x))}$, то 

$q_{i}=(1-t)y_{{i-1}}+ty_{i}+t(1-t)(a_{i}(1-t)+b_{i}t)$, где 

$a_i=k_{i-1}(x_i-x_{i-1})-(y_i - y_{i-1})$
 
$b_i=-k_i(x_i-x_{i-1})+(y_i - y_{i-1})$

Значения $k$ можно получить, решив следующую систему методом прогонки:


${\begin{bmatrix}{b_{1}}&{c_{1}}&{}&{}&{0}\\{a_{2}}&{b_{2}}&{c_{2}}&{}&{}\\{}&{a_{3}}&{b_{3}}&\ddots &{}\\{}&{}&\ddots &\ddots &{c_{n-1}}\\{0}&{}&{}&{a_{n}}&{b_{n}}\\\end{bmatrix}}{\begin{bmatrix}{k_{1}}\\{k_{2}}\\{k_{3}}\\\vdots \\{k_{n}}\\\end{bmatrix}}={\begin{bmatrix}{d_{1}}\\{d_{2}}\\{d_{3}}\\\vdots \\{d_{n}}\\\end{bmatrix}}.$

Диагональные значения можно представить как следующие векторы:

$a=\begin{bmatrix}
\frac{1}{x_{i+1}-x_i}
\end{bmatrix}$

$b = \begin{bmatrix}
\frac{2}{x_1-x_0} & 2\ \left(\frac {1}{x_1-x_0}+ \frac {1}{{x_2-x_1}}\right)  & ... & 2\ \left(\frac {1}{x_{i+1}-x_{i}}+ \frac {1}{{x_{i+2}-x_{i+1}}}\right) & ... & 2\ \left(\frac {1}{x_{n-1}-x_{n-2}}+ \frac {1}{{x_{n}-x_{n-1}}}\right) & \frac{2}{x_{n}-x_{n-1}}
\end{bmatrix}$
 
$c=\begin{bmatrix}
\frac{1}{x_{i+1}-x_i}
\end{bmatrix}$

$d=\begin{bmatrix}
3\ \frac{y_1-y_0}{(x_1-x_0)^2} &
3\ \left(\frac {y_1 - y_0}{{(x_1-x_0)}^2}+\frac {y_2 - y_1}{{(x_2-x_1)}^2}\right) &
... &
3\ \left(\frac {y_{i+1} - y_i}{{(x_{i+1}-x_i)}^2}+\frac {y_{i+2} - y_{i+1}}{{(x_{i+2}-x_{i+1})}^2}\right) &
... &
3\ \left(\frac {y_{n-1} - y_{n-2}}{{(x_{n-1}-x_{n-2})}^2}+\frac {y_{n} - y_{n-1}}{{(x_{n}-x_{n-1})}^2}\right) &
3\ \frac{y_n-y_{n-1}}{(x_n-x_{n-1})^2}
\end{bmatrix}$
 
## Метод прогонки

Метод прогонки – это упрощенный метод Гаусса, который может быть использован для решения уравнения $A*k=d$, где $A$ – трехдиагональная матрица. Тридиагональная матрица для $n$ неизвестных может быть записана следующим образом: $a_{i}*k_{i-1}+b_{i}*k_{i}+c_{i}*k_{i+1}=d_{i}$, where $a_{1}=0$ and $c_{n}=0$. 

Метод прогонки в целом нестабилен, за исключением тех случаев, когда есть диагональное преобладание, либо матрица положительно определена.

Решение такой ситемы может быть найдена следующим образом:

Сначала заменим коэффиценты $c$:

$c'_{i}={\begin{cases}{\begin{array}{lcl}{\cfrac {c_{i}}{b_{i}}}&;&i=1\\{\cfrac {c_{i}}{b_{i}-a_{i}c'_{i-1}}}&;&i=2,3,\dots ,n-1\\\end{array}}\end{cases}}\,$

Затем заменим коэффиценты $c$:

$d'_{i}={\begin{cases}{\begin{array}{lcl}{\cfrac {d_{i}}{b_{i}}}&;&i=1\\{\cfrac {d_{i}-a_{i}d'_{i-1}}{b_{i}-a_{i}c'_{i-1}}}&;&i=2,3,\dots ,n.\\\end{array}}\end{cases}}\,$

Подсчитаетм значения $k$:

$k_{n}=d'_{n}\,$

$k_{i}=d'_{i}-c'_{i}k_{i+1}\qquad ;\ i=n-1,n-2,\ldots ,1.$


## Построение графиков и тестирование алгоритма

Проинтерполируем $3$ функции (линейную функцию, синус, квадратную волну, треугольную волну). Для каждого теста нарисуем график функции и сплайна, построенного на двух сетках (одна гуще другой). Отдельно приведем график разности $E(x) = f(x) − S(x)$, где $f$ – тестовая функция, а $S$ – её интерполяция, построенная по самой густой сетке из приведённых.

```python
import numpy as np
from scipy.signal import square, sawtooth
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore") # Игнорируем предупреждения
%matplotlib inline
fig_size = [8, 6] # Настраиваем размер фигуры
plt.rcParams["figure.figsize"] = fig_size
from numerical_methods import CubicSplineInterpolation
```

```python
def linear(x):
    return x

md = CubicSplineInterpolation()

arrs = [np.arange(1, 10, 0.1), np.arange(1, 10)]
arrs_names = ['thick', 'thin']
for func in [linear, np.sin, sawtooth, square]:
    for arr in range(2):
        plt.plot(arrs[arr], func(arrs[arr]), 'o', label='data')
        md.fit(func(arrs[arr]), arrs[arr])
        yc, xc = md.predict()
        plt.plot(xc, yc, label='S')
        plt.legend(loc='upper left', ncol=2)
        plt.title(r'{} interpolation for {} array'.format(func.__name__, arrs_names[arr]))
        plt.ylabel(r'$y$')
        plt.xlabel(r'$x$')
        plt.show()

        plt.plot(xc, func(xc)-yc, "g", label='$E(x) = f(x) − S(x)$')
        plt.legend(loc='upper left', ncol=2)
        plt.title(r'{} interpolation error for {} array'.format(func.__name__, arrs_names[arr]))
        plt.ylabel(r'$E(x)$')
        plt.xlabel(r'$x$')
        plt.show()
```

## Выводы 

Эффект, который можно наблюдать на примере разницы между сплайнами разрывныx функций на разных сетках, вероятно, можно обяснить феноменом Гиббса – чем больше плотность сетки используется при реконструкции, тем ближе функция приближается к исходной форме. Чем больше будет плотность, тем ширина выброса сделается меньше, однако амплитуда выброса останется той же самой. Выброс исчезнет для плотных сеток на дискретных наборов, но не для непрерывных.

В случае интерполяции линейной функции и синуса ошибка может быть объяснена числом с плавающей точкой.
