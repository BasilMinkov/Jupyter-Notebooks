```python
import numpy as np
from scipy.stats import norm, gaussian_kde
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)

%pylab inline
```

```python
pylab.rcParams['figure.figsize'] = (14, 10)
```

```python
n = 100
xs = np.linspace(-1,1,300)
rang = np.arange(2, 30)
means = np.ones(rang.shape[0])

for m in range(rang.shape[0]):
    ar = np.array([np.mean(np.random.normal(loc=0, scale=1, size=rang[m])) for i in range(100)])
    density = gaussian_kde(ar) 
    plt.plot(xs, density(xs), label="n = {}".format(str(rang[m])))  # arguments are passed to np.histogram
    means[m] = ar.mean()
plt.legend()
plt.title(r"$KDE(\bar X_{m}). |\bar X_{m}| = 100. $", fontsize=20)
plt.ylabel(r"$KDE(\bar X_{m})$")
plt.xlabel(r"$\bar X_{m}$")
plt.show()

plt.plot(means)
plt.show()
```

```python
def s(arr, mu, ddof):
    return np.sum((arr - mu)**2)/(arr.shape[0]-ddof)
```

```python
n = 100
xs = np.linspace(0,2,300)
for m in np.arange(2, 30):
    ar = np.array([s(np.random.normal(loc=0, scale=1, size=m), 0, 0) for i in range(100)])
#     print(bar_s(np.random.normal(loc=0, scale=1, size=m), 0))
    density = gaussian_kde(ar) 
    plt.plot(xs, density(xs), label="m = {}".format(str(m)))  # arguments are passed to np.histogram
plt.legend()
plt.title(r"$ KDE(\bar S_{m}). |\bar S_{m}| = 100. $", fontsize=20)
plt.ylabel(r"$KDE(\bar S_{m})$")
plt.xlabel(r"$\bar S_{m}$")
plt.show()
```

```python
n = 100
xs = np.linspace(0,2,300)
for m in np.arange(2, 30):
    ar = np.array([s(np.random.normal(loc=0, scale=1, size=m), 0, 1) for i in range(100)])
    density = gaussian_kde(ar) 
    plt.plot(xs, density(xs), label="m = {}".format(str(m)))  # arguments are passed to np.histogram
plt.title(r"$ KDE(\tilde S_{m}). |\tilde S_{m}| = 100. $", fontsize=20)
plt.ylabel(r"$KDE(\tilde S_{m})$")
plt.xlabel(r"$\bar S_{m}$")
plt.legend()
plt.show()
```
