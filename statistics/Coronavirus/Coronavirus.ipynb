```python
import os
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

from scipy.stats import pearsonr
```

Define a function to add countres' names to scatterplot.

```python
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.03, point['y'], str(point['val']), size=10)
```

Define the path to the folder with data.

```python
path = os.path.join(os.path.abspath(os.getcwd()), "Coronavirus/")
```

Read the data to the _pandas_ dataframe.

```python
cases = pd.read_csv(path+"total_cases.csv")
adj = pd.read_csv(path+"API_NY.ADJ.NNTY.PC.CD_DS2_en_csv_v2_890678.csv")
gdp = pd.read_csv(path+"API_NY.GDP.MKTP.CD_DS2_en_csv_v2_887264.csv")
```

Subset the wages data to the last year, that for which the largest amount of information is avaliable.

```python
adj = adj.loc[:, ["Country Name", "2017"]].dropna()
adj = adj.rename(columns={"2017": "Average Wages"})
```

Add some data on Russian average wage.

```python
adj = adj.append(pd.DataFrame({"Country Name": ["Russia"], "Average Wages": [6511]}))
```

Subset coronavirus data to use only the 20th of March comulative info. 

```python
df = cases.iloc[80, 2:].reset_index()
df.columns = ["Country Name", "COVID-19 Cases"]
df = pd.merge(df, adj, on="Country Name")
```

Delete countries for which we don't have any data ether on case or on wage.

```python
df = df.dropna()
```

Delete some countries to avoid intersection of the dotes on the final scatterplot. Note, that we will anyway use the data on these countries for the statistical test.

```python
dfx = df[~df["Country Name"].isin(
    [
        "Liberia", 
        "Qatar", 
        "Luxembourg", 
        "Jordan", 
        "South Africa", 
        "Ecuador",
        "Burkina Faso", 
        "Panama", 
        "Estonia"
    ]
)]
```

Set up _seaborn_ graphics and make resulting scatterplot.

```python
# set up _seaborn_ graphics
sns.set()
plt.rcParams.update({'font.size': 22})

# here we define figure and use regplot function from seaborn to make scatterplot with the reggression line
f, ax = plt.subplots(figsize=(30, 30))
sns.regplot(
    x=np.log(dfx['COVID-19 Cases'].astype("float")), 
    y=np.log(dfx["Average Wages"].astype("float")), 
    ax=ax,
    ci=None
)

# here we plot lables with the previously defined function
label_point(
    np.log(dfx['COVID-19 Cases'].astype("float")), 
    np.log(dfx["Average Wages"].astype("float")), 
    dfx["Country Name"], 
    ax
)  

# calculate Pearson test
test = pearsonr(np.log(df['COVID-19 Cases'].astype("float")), np.log(df["Average Wages"].astype("float")))

# add title and axis names
plt.title("Average Wage And Cumulative COVID-17 Cases Scatter Plot\nPearson Correlation = {}; p-value < 0.001".format(round(test[0], 2)), 
          size=35,
         weight="bold"
         )
plt.ylabel("Logarithm Of Annual Average Wages In 2017", size=30)
plt.xlabel("Logarithm Of Cumulative COVID-19 Cases By 20 May 2020", size=30)
plt.savefig("my_plot.png", dpi=100)
```
